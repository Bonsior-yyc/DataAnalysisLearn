```python
#matplotlib 学习

import matplotlib.pyplot as plt
import numpy as np
a = np.linspace(1,10,10).flatten()
b = np.array([1,3,5,6,3,4,2,7,4])
l1, = plt.plot(a,color="red",linestyle="--",label="lbr")
l2, = plt.plot(b,color="black",label="sb")
#图例
plt.legend(handles=[l1,l2,])
#坐标轴标志
plt.xlabel("lbrsb")
plt.ylabel("lbrsbbb")
plt.legend
#切换图像
plt.figure(2)
x = np.linspace(-10,10,100).flatten()
y = x**2 + 1
plt.plot(x,y)
#控制显示范围
plt.xlim((-5,5))
plt.ylim((10,100))
#切换坐标轴标
new_tick = np.linspace(0,100,20)
plt.yticks(new_tick)
new_tick1 = [-10,0,10]
new_tick2 = ["l","b","r"]
plt.xticks(new_tick1,new_tick2)
#坐标轴位置
ax = plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.xaxis.set_ticks_position("bottom")
ax.yaxis.set_ticks_position("left")
ax.spines["bottom"].set_position(("data",0))
ax.spines["left"].set_position(("data",0))
#注解
plt.figure(3)
x1 = np.linspace(-10,10,20)
y1 = 2*x1 + 1
plt.plot(x1,y1)
ax1 = plt.gca()
ax1.spines["right"].set_color("none")
ax1.spines["top"].set_color("none")
ax1.xaxis.set_ticks_position("bottom")
ax1.yaxis.set_ticks_position("left")
ax1.spines["bottom"].set_position(("data",0))
ax1.spines["left"].set_position(("data",0))
x0 = 1
y0 = 2*x0 + 1
plt.scatter(x0,y0,color="red")
plt.annotate("lbrissb",xy=(x0+1,y0-2))
plt.annotate("lbrsb",xy=(7.5,-15),size="20")
plt.show()
```
